## Условие 3: 3 задача
#### Составьте несколько сводных таблиц

#### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален. 
#### Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values(by='price', ascending=False)
```

#### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['min', 'max', 'mean']})
```

#### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.groupby(['view', 'waterfront']).count().id
```

```python
df.pivot_table(
    index= 'view',
    columns= 'waterfront',
    aggfunc= 'count',
    values= 'id',
    fill_value= '-'
)
```

#### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
df.groupby(['floors', 'bedrooms']).count().id
```

```python
df.pivot_table(
    index= 'floors',
    columns= 'bedrooms',
    aggfunc= 'count',
    values= 'id',
    fill_value= '-'
)
```

#### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df['price'].median()
```

```python
df.groupby(['condition', 'grade']).median().price
```

```python
df.pivot_table(
    index= 'condition',
    columns= 'grade',
    aggfunc= 'median',
    values= 'price',
    fill_value= '-'
)
```
